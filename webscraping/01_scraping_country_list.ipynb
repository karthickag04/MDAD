```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl
```

```python

# The URL of the webpage
url = "https://catking.in/list-of-countries-capitals-currencies-of-the-world"

# Send a GET request to fetch the HTML content
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table by the tag <table>
table = soup.find('table')
```

```python
print(table)
```

```python

# Extract the headers from the first row (the <tr> tag within <thead> or first <tr>)
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())
print(headers)

# If the headers are within a <tr> inside <tbody>, use this:
if not headers:
    headers = [header.text.strip() for header in table.find_all('td')[:4]]  # Modify according to your needs
print(headers)
# # Extract the rows
rows = []
for row in table.find_all('tr')[1:]:  # Skipping the header row
    cols = row.find_all('td')
    # print(cols)
    cols = [element.text.strip() for element in cols]
    # print (cols)
    rows.append(cols)

print(rows)
```

```python
# Convert the data into a Pandas DataFrame
df = pd.DataFrame(rows, columns=headers)
df.head()
```

```python
# Display the DataFrame
print(df)

# Optionally, save the DataFrame to a CSV file
df.to_csv('countries_capitals_currencies.csv', index=False)
```
