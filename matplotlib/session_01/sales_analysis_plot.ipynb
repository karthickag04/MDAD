# Sales Analysis
This notebook provides a comprehensive sales analysis using the dataset `dataset_supermarket_sales.csv`.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Load the dataset
df = pd.read_csv('dataset_supermarket_sales.csv')

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# Extract hour from 'Time'
df['Hour'] = df['Time'].apply(lambda x: int(x.split(':')[0]))
```

## 1. Descriptive Statistics

```python
# Total Sales
total_sales = df['Total'].sum()
total_sales
```

```python
# Average Sales
average_sales = df['Total'].mean()
average_sales
```

### Sales by Branch

```python
# Sales by Branch
sales_by_branch = df.groupby('Branch')['Total'].sum()

sales_by_branch.plot(kind='bar', title='Sales by Branch')
plt.ylabel('Total Sales')
plt.show()
sales_by_branch
```

### Sales by City

```python
# Sales by City
sales_by_city = df.groupby('City')['Total'].sum()
sales_by_city.plot(kind='bar', title='Sales by City')
plt.ylabel('Total Sales')
plt.show()
sales_by_city
```

### Sales by Customer Type

```python
# Sales by Customer Type
sales_by_customer_type = df.groupby('Customer type')['Total'].sum()
sales_by_customer_type.plot(kind='bar', title='Sales by Customer Type')
plt.ylabel('Total Sales')
plt.show()
sales_by_customer_type
```

### Sales by Gender

```python
# Sales by Gender
sales_by_gender = df.groupby('Gender')['Total'].sum()
sales_by_gender.plot(kind='bar', title='Sales by Gender')
plt.ylabel('Total Sales')
plt.show()
sales_by_gender
```

### Sales by Product Line

```python
# Sales by Product Line
sales_by_product_line = df.groupby('Product line')['Total'].sum()
sales_by_product_line.plot(kind='bar', title='Sales by Product Line')
plt.ylabel('Total Sales')
plt.show()
sales_by_product_line
```

### Sales by Payment Method

```python
# Sales by Payment Method
sales_by_payment = df.groupby('Payment')['Total'].sum()
sales_by_payment.plot(kind='bar', title='Sales by Payment Method')
plt.ylabel('Total Sales')
plt.show()
sales_by_payment
```

### Average Rating by Product Line

```python
# Average Rating by Product Line
average_rating_by_product_line = df.groupby('Product line')['Rating'].mean()
average_rating_by_product_line.plot(kind='bar', title='Average Rating by Product Line')
plt.ylabel('Average Rating')
plt.show()
average_rating_by_product_line
```

## 2. Time Series Analysis

### Monthly Sales Trend

```python
# Monthly Sales Trend
monthly_sales_trend = df.resample('M', on='Date')['Total'].sum()
monthly_sales_trend.plot(kind='line', title='Monthly Sales Trend')
plt.ylabel('Total Sales')
plt.show()
monthly_sales_trend
```

### Daily Sales Trend

```python
# Daily Sales Trend
daily_sales_trend = df.resample('D', on='Date')['Total'].sum()
daily_sales_trend.plot(kind='line', title='Daily Sales Trend')
plt.ylabel('Total Sales')
plt.show()
daily_sales_trend
```

### Sales by Hour of the Day

```python
# Sales by Hour
sales_by_hour = df.groupby('Hour')['Total'].sum()
sales_by_hour.plot(kind='bar', title='Sales by Hour of the Day')
plt.ylabel('Total Sales')
plt.show()
sales_by_hour
```

## 4. Product Analysis

### Most Popular Products

```python
# Most Popular Products
most_popular_products = df.groupby('Product line')['Quantity'].sum().sort_values(ascending=False)
most_popular_products.plot(kind='bar', title='Most Popular Products')
plt.ylabel('Quantity Sold')
plt.show()
most_popular_products
```

### Least Popular Products

```python
# Least Popular Products
least_popular_products = df.groupby('Product line')['Quantity'].sum().sort_values(ascending=True)
least_popular_products.plot(kind='bar', title='Least Popular Products')
plt.ylabel('Quantity Sold')
plt.show()
least_popular_products
```

### Revenue Contribution by Product Line

```python
# Revenue Contribution
total_sales = df['Total'].sum()
revenue_contribution = df.groupby('Product line')['Total'].sum() / total_sales * 100
revenue_contribution.plot(kind='bar', title='Revenue Contribution by Product Line')
plt.ylabel('Percentage Contribution')
plt.show()
revenue_contribution
```

## 6. Rating Analysis

### Average Rating by Product Line

```python
# Average Rating by Product Line
average_rating_by_product_line = df.groupby('Product line')['Rating'].mean()
average_rating_by_product_line.plot(kind='bar', title='Average Rating by Product Line')
plt.ylabel('Average Rating')
plt.show()
average_rating_by_product_line
```

### Rating Distribution

```python
# Rating Distribution
rating_distribution = df['Rating'].value_counts().sort_index()
rating_distribution.plot(kind='bar', title='Rating Distribution')
plt.ylabel('Frequency')
plt.xlabel('Rating')
plt.show()
rating_distribution
```

```python
# Round ratings to the nearest whole number (1 to 10)
df['Rounded Rating'] = np.round(df['Rating']).clip(lower=1, upper=10)

# Rating Distribution
rating_distribution = df['Rounded Rating'].value_counts().sort_index()

# Plotting
rating_distribution.plot(kind='bar', title='Rating Distribution')
plt.ylabel('Frequency')
plt.xlabel('Rating')
plt.show()

# Display rating distribution
rating_distribution
```

## 7. Branch Analysis

### Branch Performance

```python
# Branch Performance
branch_performance = df.groupby('Branch').agg({'Total': ['sum', 'mean', 'count']})
branch_performance
```

## 8. Profit Analysis

### Gross Profit and Profit Margin

```python
# Gross Profit
df['Gross Profit'] = df['Total'] - df['cogs']
# Profit Margin
df['Profit Margin'] = (df['Gross Profit'] / df['Total']) * 100
gross_profit = df['Gross Profit'].sum()
profit_margin = df['Profit Margin'].mean()
gross_profit, profit_margin
```

```python
# Profit Margin by Product Line
profit_margin_by_product_line = df.groupby('Product line')['Profit Margin'].mean()
profit_margin_by_product_line.plot(kind='bar', title='Profit Margin by Product Line')
plt.ylabel('Average Profit Margin (%)')
plt.show()
profit_margin_by_product_line
```
